```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, ElasticNet, Lasso, BayesianRidge, LassoLarsIC
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
import xgboost
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV

import warnings
def ignore_warn(*args, **kwargs):
    pass
warnings.warn = ignore_warn #ignore annoying warning (from sklearn and seaborn)
```

```python
train = pd.read_csv('../data/train_transformed.csv')
test = pd.read_csv('../data/test_transformed.csv')
```

## Build Models

```python
y_tr = np.log(1 + train['SalePrice'])
x_tr = train.drop('SalePrice', axis=1)
```

```python
def cv_mean_mse(model, n_folds=10):
    mean_mse = -cross_val_score(model, x_tr.values, y_tr, cv = n_folds, \
                                scoring="neg_mean_squared_error").mean()
    return(mean_mse)
```

Compute c.v. MSE and best alpha

```python
alpha_ri_li = [0.1*i for i in range(100)] 
mse_ri_li = list()

for i in alpha_ri_li:
    mse_ri_li.append(cv_mean_mse(Ridge(alpha = i)))
    
best_alpha_ri = alpha_ri_li[mse_ri_li.index(min(mse_ri_li))] # select the best alpha
```

```python
alpha_la_li = [0.0001*i for i in range(100)] 
mse_la_li = list()

for i in alpha_la_li:
    mse_la_li.append(cv_mean_mse(Lasso(alpha = i)))
    
best_alpha_la = alpha_la_li[mse_la_li.index(min(mse_la_li))]
```

Compute training MSE

```python
# ridge
ri_model = Ridge(best_alpha_ri).fit(x_tr, y_tr)
y_hat_tr_ri = ri_model.predict(x_tr)
ri_tr_mse = mean_squared_error(y_hat_tr_ri, y_tr)
```

```python
# lasso
la_model = Lasso(best_alpha_la).fit(x_tr, y_tr)
y_hat_tr_la = la_model.predict(x_tr)
la_tr_mse = mean_squared_error(y_hat_tr_la, y_tr)
```

```python
print('Ridge: training MSE: %f, c.v. MSE: %f, best alpha: %f.' % (ri_tr_mse, min(mse_ri_li), best_alpha_ri))
print('LASSO: training MSE: %f, c.v. MSE: %f, best alpha: %f.' % (la_tr_mse, min(mse_la_li), best_alpha_la))
```

### Kernel ridge regression 

```python
# kernel ridge regression 

coef0_li = [i for i in range(-5, 5)]
alpha_krr_li = [i/100 for i in range(0, 10)]
parameters = {'alpha': alpha_krr_li, 'coef0': coef0_li}

gscv_krr = GridSearchCV(KernelRidge(kernel='polynomial', degree=2), parameters, cv=2, scoring = "neg_mean_squared_error")
gscv_krr.fit(x_tr, y_tr)
print(gscv_krr.best_params_)
print(gscv_krr.best_score_)
```

```python
coef0_li = [i for i in range(10000, 10005)]
alpha_krr_li = [i/100 for i in range(10000, 10005)]
parameters = {'alpha': alpha_krr_li, 'coef0': coef0_li}

gscv_krr = GridSearchCV(KernelRidge(kernel='polynomial', degree=2), parameters, cv=2, scoring = "neg_mean_squared_error")
gscv_krr.fit(x_tr, y_tr)
print(gscv_krr.best_params_)
print(gscv_krr.best_score_)
```

```python
mse_krr = cv_mean_mse(KernelRidge(kernel='polynomial', degree=2))
krr_model = KernelRidge(kernel='polynomial', degree=2).fit(x_tr, y_tr)

y_hat_tr_krr = krr_model.predict(x_tr)
krr_tr_mse = mean_squared_error(y_hat_tr_krr, y_tr) 

print('Kernel ridge regression: training MSE: %f, c.v. MSE: %f.' % (krr_tr_mse, mse_krr))
```

```python
mse_krr = cv_mean_mse(KernelRidge(kernel='polynomial', degree=2, alpha=10000000, coef0=10000000))
krr_model = KernelRidge(kernel='polynomial', degree=2).fit(x_tr, y_tr)

y_hat_tr_krr = krr_model.predict(x_tr)
krr_tr_mse = mean_squared_error(y_hat_tr_krr, y_tr) 

print('Kernel ridge regression: training MSE: %f, c.v. MSE: %f.' % (krr_tr_mse, mse_krr))
```

```python
# Gradient Boosting Regression

mse_gbr = cv_mean_mse(GradientBoostingRegressor(loss='huber'))
gbr = GradientBoostingRegressor(loss='huber')
gbr_model = gbr.fit(x_tr, y_tr)

y_hat_tr_gbr = gbr_model.predict(x_tr)
gbr_tr_mse = mean_squared_error(y_hat_tr_gbr, y_tr)
```

```python
print('Gradient Boosting Regression: training MSE: %f, c.v. MSE: %f.' % (gbr_tr_mse, mse_gbr))
```

```python
mse_xgb = cv_mean_mse(xgboost.XGBRegressor())
xgb = xgboost.XGBRegressor()
xgb_model = xgb.fit(x_tr, y_tr)

y_hat_tr_xgb = xgb_model.predict(x_tr)
xgb_tr_mse = mean_squared_error(y_hat_tr_xgb, y_tr)
```

```python
print('XGBoost: training MSE: %f, c.v. MSE: %f.' % (xgb_tr_mse, mse_xgb))
```
